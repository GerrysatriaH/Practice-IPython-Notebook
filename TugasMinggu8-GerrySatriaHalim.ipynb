## Soal 2

```python
import cv2 as cv   # Mengimport library OpenCV
import numpy as np # Mengimport library Numpy
import requests    # Mengimport requests

# Membaca gambar dalam mode grayscale
img = cv.imdecode(np.frombuffer(requests.get("https://raw.githubusercontent.com/promecarus/sandboxMultipleLanguage/main/img/gradient.png").content, np.uint8), cv.IMREAD_GRAYSCALE) 

_, th1 = cv.threshold(img, 50, 255, cv.THRESH_BINARY)      # Menerapkan threshold binary pada gambar dengan nilai ambang 50
_, th2 = cv.threshold(img, 200, 255, cv.THRESH_BINARY_INV) # Menerapkan threshold binary inverse pada gambar dengan nilai ambang 200
_, th3 = cv.threshold(img, 127, 255, cv.THRESH_TRUNC)      # Menerapkan threshold truncate pada gambar dengan nilai ambang 127
_, th4 = cv.threshold(img, 127, 255, cv.THRESH_TOZERO)     # Menerapkan threshold tozero pada gambar dengan nilai ambang 127
_, th5 = cv.threshold(img, 127, 255, cv.THRESH_TOZERO_INV) # Menerapkan threshold tozero inverse pada gambar dengan nilai ambang 127


cv.imshow("Image", img) # Menampilkan gambar asli
cv.imshow("th1", th1)   # Menampilkan hasil gambar yang diterapkan threshold binary
cv.imshow("th2", th2)   # Menampilkan hasil gambar yang diterapkan threshold binary inverse
cv.imshow("th3", th3)   # Menampilkan hasil gambar yang diterapkan threshold truncate
cv.imshow("th4", th4)   # Menampilkan hasil gambar yang diterapkan threshold tozero
cv.imshow("th5", th5)   # Menampilkan hasil gambar yang diterapkan threshold tozero inverse

cv.waitKey(0)          # Menunggu pengguna untuk menekan tombol keyboard
cv.destroyAllWindows() # Menutup semua jendela tampilan gambar 
```

## Soal 3

```python
import cv2 as cv   # Mengimport library OpenCV
import numpy as np # Mengimport library Numpy
from matplotlib import pyplot as plt   # Mengimport library matplotlib
import requests    # Mengimport requests

img = cv.imread('images/sudoku.png',0) # Membaca gambar dalam mode grayscale
img = cv.medianBlur(img, 5)            # Menerapkan median filter pada gambar dengan kernel ukuran 5x5.

# Menerapkan threshold binary pada gambar dengan nilai ambang 127
ret,th1 = cv.threshold(img,127,255,cv.THRESH_BINARY)  

# Menerapkan thresholding adaptif pada gambar dengan menggunakan metode ADAPTIVE_THRESH_MEAN_C
th2 = cv.adaptiveThreshold(img,255,cv.ADAPTIVE_THRESH_MEAN_C,\
                           cv.THRESH_BINARY,11,2)     

# Menerapkan thresholding adaptif pada gambar dengan menggunakan metode ADAPTIVE_THRESH_GAUSSIAN_C
th3 = cv.adaptiveThreshold(img,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,\
                           cv.THRESH_BINARY,11,2)     

# Menyimpan daftar judul pada array
title = ['Original Image', 'Global Thresholding (v = 127)',
         'Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']

# Menyimpan hasil gambar threshold
images = [img, th1, th2, th3]

for i in range(4) : # Melakukan looping sebanyak 4 kali, karena ada 4 citra yang akan ditampilkan.
    plt.subplot(2,2,i+1) # Membuat grid plot 2x2 dan menempatkan citra ke-i pada grid tersebut.
    plt.imshow(images[i], 'gray') # Menampilkan citra ke-i dalam format grayscale.
    plt.title(title[i])  # Memberikan judul pada citra ke-i sesuai dengan daftar judul yang sudah dibuat.
    plt.xticks([]) # Menghilangkan nilai sumbu x pada plot.
    plt.yticks([]) # Menghilangkan nilai sumbu y pada plot.
    
plt.show() # Menampilkan Plot
```
